# Heatmap

```python
plt.rc('font', family='applegothic')
```

```python
corr = df2.corr()
```

```python
corr
```

```python
mask = np.zeros_like(corr, dtype=np.bool_)
mask[np.triu_indices_from(mask)] = True
```

```python
import matplotlib as mpl
print(mpl.matplotlib_fname())
```

```python
sns.heatmap(corr, 
            annot = True,      # 실제 값 화면에 나타내기
            cmap = 'RdYlBu_r',  # Red, Yellow, Blue 색상으로 표시
            vmin = -1, vmax = 1, #컬러차트 -1 ~ 1 범위로 표시
            )
```

```python
# 상관계수가 높은 순서대로 정렬
sorted_corr = corr.unstack().sort_values(ascending=False)

# 중복된 쌍 제거 (좌우 반전된 쌍은 중복으로 처리)
unique_sorted_corr = sorted_corr[sorted_corr.index.map(lambda x: x[0] < x[1])]
```

```python
df3 = unique_sorted_corr.to_frame().reset_index()
df3.columns = ['var1','var2','corr']
df3[df3['var1']=='경영_영업이익']
```
