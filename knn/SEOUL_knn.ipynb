```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
seoul_class1 = pd.read_csv('../data/SEOUL_CLASS_1.csv')
```

```python
seoul_class1.drop(['기준_년_코드', '기준_분기_코드', '상권_코드', '상권_코드_명'], axis=1, inplace=True)
seoul_class1
```

```python
cluster0_df = seoul_class1[seoul_class1['클러스터'] == 0].copy()
cluster1_df = seoul_class1[seoul_class1['클러스터'] == 1].copy()
cluster2_df = seoul_class1[seoul_class1['클러스터'] == 2].copy()
cluster3_df = seoul_class1[seoul_class1['클러스터'] == 3].copy()
```

```python
cluster0_df.drop('클러스터', axis=1, inplace=True)
cluster1_df.drop('클러스터', axis=1, inplace=True)
cluster2_df.drop('클러스터', axis=1, inplace=True)
cluster3_df.drop('클러스터', axis=1, inplace=True)
```

### 1) 클러스터 0

```python
x1 = cluster0_df[cluster0_df['업종_대분류'] == '외식업'].drop(['경영_위기_비율', '업종_대분류'], axis=1)
y1 = cluster0_df[cluster0_df['업종_대분류'] == '외식업']['경영_위기_비율']
x1_train, x1_valid, y1_train, y1_valid = train_test_split(x1, y1, test_size=0.2, random_state=0)
```

```python
k = 3
knn_regressor = KNeighborsRegressor(n_neighbors=k)
knn_regressor.fit(x1_train, y1_train)
```

```python
y_pred = knn_regressor.predict(x1_valid)
```

```python
mse = mean_squared_error(y1_valid, y_pred)
```
