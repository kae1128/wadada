# 데이터 정리 기준 불러오기
- 2022년 1월 ~ 12월

```python
# 데이터 정의서 파일 경로 입력
상권기준 = pd.read_excel(directory_path+'2023빅콘테스트_데이터신기술분야_데이터정의서.xlsx',sheet_name = 2).iloc[2:,:]
상권기준.columns=['상권_코드_명','상권_코드']
상권기준 = 상권기준.reset_index(drop=True)
상권기준
```

```python
# 필요한 데이터 dict로 넣음. 파일명 필요해서 list 말고 dict로 ..
data_dict = {'점포':점포,
            '추정매출':추정매출,
            '상권변화지표':상권변화지표,
            '집객시설':집객시설,
            '아파트':아파트,
            '상주인구':상주인구,
            '직장인구':직장인구,
            '생활인구':생활인구}
```

```python
# key 컬럼 중 컬럼명 다른 애들 변경해주기
상주인구.rename(columns={'상권 코드 명':'상권_코드_명',
                         '상권 코드':'상권_코드'},inplace=True)
직장인구.rename(columns={'기준_년월_코드':'기준_년_코드'},inplace=True)
생활인구.rename(columns={'기준 년코드':'기준_년_코드'},inplace=True)
```

```python
processed_data_dict = {}

for df_name,df in data_dict.items():
    # 상권 기준 안에 있는 상권들만 추출
    df_filtered = df[df['상권_코드'].isin(상권기준['상권_코드'])]
    # 2022년 데이터만 추출
    df_filtered = df_filtered[df_filtered['기준_년_코드'] == 2022]
    processed_data_dict[df_name] = df_filtered  # Store processed data
    print('Name of the district : ',df_name)
    print('Only 2022 data included : ',df_filtered['기준_년_코드'].nunique()==1)
    print('All quarters included : ',df_filtered['기준_분기_코드'].nunique()==4)
    # 그 결과로 몇 개 상권이 뽑혔는지 확인
    print('Number of districts : ',df_filtered['상권_코드_명'].nunique(),'/ 113')
    print('---------------------------------------------')
```

```python
# dict 안에서 변경된 df들을 다시 밖으로 끄집어내기 ..
점포 = processed_data_dict['점포']
추정매출 = processed_data_dict['추정매출']
상권변화지표 = processed_data_dict['상권변화지표']
집객시설 = processed_data_dict['집객시설']
아파트 = processed_data_dict['아파트']
상주인구 = processed_data_dict['상주인구']
직장인구 = processed_data_dict['직장인구']
생활인구 = processed_data_dict['생활인구']
```

```python
# 모든 상권, 2022 1~4분기를 포함한 데이터 프레임 생성 = base_df
# 여기다가 다른 상권들 left로 merge
base_df.rename(columns={'trdar_nm':'상권_코드_명','quarter':'기준_분기_코드'},inplace=True)
base_df = base_df[['상권_코드_명','기준_분기_코드']]
base_df
```

```python
a = pd.merge(base_df, 점포, on=['상권_코드_명','기준_분기_코드'],how='outer')
a = pd.merge(a, 추정매출, on=['상권_코드_명','기준_분기_코드'],how='outer')
a = pd.merge(a, 상권변화지표, on=['상권_코드_명','기준_분기_코드'],how='outer')
a = pd.merge(a, 집객시설, on=['상권_코드_명','기준_분기_코드'],how='outer')
a = pd.merge(a, 아파트, on=['상권_코드_명','기준_분기_코드'],how='outer')
a = pd.merge(a, 상주인구, on=['상권_코드_명','기준_분기_코드'],how='outer')
a = pd.merge(a, 직장인구, on=['상권_코드_명','기준_분기_코드'],how='outer')
a = pd.merge(a, 생활인구, on=['상권_코드_명','기준_분기_코드'],how='outer')

a
```
