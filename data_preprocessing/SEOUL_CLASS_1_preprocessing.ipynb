```python
import pandas as pd
```

```python
class1 = pd.read_csv('../data/CREDIT_CLASS_1.csv')
class1
```

```python
class1.columns
```

```python
class1_isrisky = class1[['trdar_nm', 'year', 'quarter', 'class_2_name', 'average(is_risky)']]
class1_isrisky.columns = ['상권_코드_명', '기준_년_코드', '기준_분기_코드', '업종_대분류', '경영_위기_비율']
```

```python
seoul = pd.read_csv('../data/SEOUL.csv')
seoul.drop('경영_위기_비율', axis=1, inplace=True)
seoul
```

```python
SEOUL_CLASS_1 = pd.merge(class1_isrisky, seoul, on=['상권_코드_명', '기준_년_코드', '기준_분기_코드'], how='left')
```
