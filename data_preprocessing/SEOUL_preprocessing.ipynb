```python
import pandas as pd
```

```python
df = pd.read_csv('../data/SEOUL.csv')
df
```

```python
df.columns
```

```python
df['총_매출_금액'] = df['주중_매출_금액'] + df['주말_매출_금액']
df['총_매출_금액_성별'] = df['남성_매출_금액'] + df['여성_매출_금액']
df['총_매출_금액_연령대'] = df['연령대_10_매출_금액']+df['연령대_20_매출_금액']+df['연령대_30_매출_금액']+df['연령대_40_매출_금액']+df['연령대_50_매출_금액']+df['연령대_60_이상_매출_금액']
```

```python
df['주중_매출_비율'] = df['주중_매출_금액'] / df['총_매출_금액']
df['주말_매출_비율'] = df['주말_매출_금액'] / df['총_매출_금액']
df['남성_매출_비율'] = df['남성_매출_금액'] / df['총_매출_금액_성별']
df['여성_매출_비율'] = df['여성_매출_금액'] / df['총_매출_금액_성별']
df['연령대_1020_매출_비율'] = (df['연령대_10_매출_금액']+df['연령대_20_매출_금액']) / df['총_매출_금액_연령대']
df['연령대_3040_매출_비율'] = (df['연령대_30_매출_금액']+df['연령대_40_매출_금액']) / df['총_매출_금액_연령대']
df['연령대_5060_매출_비율'] = (df['연령대_50_매출_금액']+df['연령대_60_이상_매출_금액']) / df['총_매출_금액_연령대']
```
