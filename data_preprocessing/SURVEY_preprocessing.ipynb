```python
df_notnull['경영_고객결재방법_합계비율']
```

```python
# null값이 있는 컬럼 정제 전인 df로 돌아가서 확인
columns_to_replace_nan = ['경영_고객결재방법_현금비율',
'경영_고객결재방법_카드비율',
'경영_고객결재방법_간편결제비율',
'경영_고객결재방법_어음비율',
'경영_고객결재방법_기타비율',
'경영_고객결재방법_합계비율']

df[columns_to_replace_nan]
```

```python
# 합계비율 열의 값이 다른 열들의 합과 일치하는지 확인하고, 일치하면 NaN 값을 0으로 대체 --> 이 코드 어떻게 지지고 볶아도 안먹음 이유가 뭐냐ㅠ
total_sum = df[columns_to_replace_nan[:-1]].sum(axis=1)
'''is_match = total_sum == df['경영_고객결재방법_합계비율']
df[columns_to_replace_nan] = df[columns_to_replace_nan].where(is_match, 0)'''

# 새롭게 해결해보아요 
# 각 행별 컬럼 합계를 리스트로 생성
ts = total_sum.to_list()

# counter 함수 활용해서 unique한 값과 그 개수 세기
# 모든 행의 합계가 100인지 확인 위함 -> 그냥 조건문 없이 Nan값 0으로 싹 대체해버리려고요
from collections import Counter
count = Counter(ts)

for value, frequency in count.items():
    print(f'값: {value}, 개수: {frequency}')
```

```python
# Nan값 0으로 변경하고 다시 확인
df[columns_to_replace_nan] = df[columns_to_replace_nan].fillna(0)
df[columns_to_replace_nan]
```

```python
df['일반_창업횟수'].value_counts()
```

```python
# 그룹화된 데이터프레임 생성
grouped_means = df.groupby('일반_창업횟수')[columns_to_replace_nan].mean()

# 그룹화된 데이터프레임과 원본 데이터프레임 병합
df = df.merge(grouped_means, left_on='일반_창업횟수', right_index=True, suffixes=('', '_mean'))

# NaN 값을 평균 값으로 대체
for col in columns_to_replace_nan:
    df[col].fillna(df[col + '_mean'], inplace=True)

# 중복 열 삭제
df.drop(columns=[col + '_mean' for col in columns_to_replace_nan], inplace=True)
```
